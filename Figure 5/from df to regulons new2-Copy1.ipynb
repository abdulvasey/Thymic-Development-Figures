```python
#finding all regulons that target FEZF2 and AIRE
list1=['FEZF2','AIRE']
lst = col_list
targs = []
tf = []
for x in lst:
    my_list = df2.loc[x][0]
    t = []
    length = 0
    for z in my_list:
        if z in list1 and z != x:
            t = t + [z]
            length = length + 1
    targs = targs + t
    nam = [x]*length
    tf = tf + nam
    #targs = targs + my_list
    #length = len(my_list)
    #nam = [x]*length
    #tf = tf + nam
    
data = {'TFs': tf,
        'targs': targs}
  
# Create DataFrame
daf = pd.DataFrame(data)
```
